```python
final_dict = {}
for i in range(3000000, 6248205, 100000):
    for j in range(7600000, 13801502, 100000):
        final_dict[(i, j)]="NA"
```

```python
for key in final_dict.keys():
    min_dist = "NA"
    min_fips = "NA"
    x = key[0]
    y = key[1]*-1
    for county_dict_key in final_county_dict.keys():
        county_geo = final_county_dict[county_dict_key]
        curr_dist = (x-county_geo[0])**2+(y-county_geo[1])**2
        if (min_dist=="NA" or curr_dist<min_dist):
            min_dist = curr_dist
            min_fips = county_dict_key
    final_dict[key]=min_fips
```

```python
final_rt = []
with open('./final_stkde.txt') as fp:
    for cnt, line in enumerate(fp):
        data = line.split(",")
        x = int(float(data[0]))
        y = int(float(data[1]))
        t = int(data[2])
        date = "NA"
        if (t<11):
            date = "2020-1-"+str(t+21)
        elif (t<40):
            date = "2020-2-"+str(t-10)
        elif (t<71):
            date = "2020-3-"+str(t-39)
        else:
            date = "2020-4-"+str(t-70)
        
        v = str(data[3])
        fips = final_dict[(x, y)]
        final_rt.append((fips, t, date, v))
```

```python
f = open("correlation_stkde.txt", "w")
for i in range(0, len(final_rt)):
    line = final_rt[i]
    curr = line[0]+","+str(line[1])+","+line[2]+","+line[3]
    f.write(curr)
f.close()
```
